```python
from sklearn.datasets import load_wine
```

```python
data = load_wine()
```

```python
X = data["data"]
y = data["target"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15, random_state=40)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    "criterion":["gini", "entropy", "log_loss"],
    'max_depth':[2,3,4,5,6],
    'min_samples_split' : [2,3,4,5],
    'min_samples_leaf' : [2,3,4],
    "max_features": ["sqrt", "log2"],
    "max_leaf_nodes" : [5,10,100]
}
```

```python
randomcv_dt = RandomizedSearchCV(DecisionTreeClassifier(), param_grid)
```

```python
randomcv_dt.fit(X_train, y_train)
```

```python
best_dt = randomcv_dt.best_estimator_
```

```python
randomcv_dt.best_params_
```

```python
y_pred = best_dt.predict(X_test)
```

```python
from sklearn.metrics import  accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

## Grow a random forest 

```python
from sklearn.model_selection import ShuffleSplit
import numpy as np
```

```python
shuffle_split = ShuffleSplit(n_splits=10, test_size=0.1, random_state=10)
```

```python
X_train = []
X_test= []
y_train = []
y_test = []
```

```python
for i, (train_index, test_index) in enumerate(shuffle_split.split(X)):
    # print(train_index)
    X_train.append(np.array([X[k] for k in train_index ]))
    X_test.append(np.array([X[k] for k in test_index ]))
    y_train.append(np.array([y[k] for k in train_index ]))
    y_test.append(np.array([y[k] for k in test_index ]))


```

```python
Dt_1 = DecisionTreeClassifier(min_samples_split= 3,
 min_samples_leaf= 2,
 max_leaf_nodes= 100,
 max_features= 'sqrt',
 max_depth= 5,
 criterion= 'gini')
```

```python
list_of_tree = []
for i in range(10):
    list_of_tree.append(DecisionTreeClassifier(min_samples_split= 3,
 min_samples_leaf= 2,
 max_leaf_nodes= 100,
 max_features= 'sqrt',
 max_depth= 5,
 criterion= 'gini'))
```

```python
### training the random forrest

for i in range(10):
    list_of_tree[i].fit(X_train[i], y_train[i])
```

```python
## prediction 

prediction = []
for i in  range(10):
    prediction.append(list_of_tree[i].predict(X_test[2]))
```

```python
prediction  = np.array(prediction)
```

```python
import pandas as pd
pred = pd.DataFrame(prediction)
```

```python
pred.co
```

```python
final_pred = []
for col in pred:
    final_pred.append(pred[col].value_counts().idxmax())
```

```python
final_pred
```

```python
accuracy_score(y_test[2],final_pred)
```
